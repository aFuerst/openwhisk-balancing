```python
import os, pickle, random
from collections import defaultdict
from math import floor, ceil
import pandas as pd
import numpy as np
import scipy
import scipy.stats 
import matplotlib as mpl
mpl.rcParams.update({'font.size': 14})
import matplotlib.pyplot as plt
```

```python
loadspace = np.linspace(0, 10, 1000) 
transformed_load = scipy.stats.norm.cdf(loadspace, loc=0.5, scale=0.1)
plt.xlim(0,1)
plt.plot(loadspace, transformed_load)
```

```python
lim = 90
for multi in [1, 1.5, 2]:
  loadspace = np.linspace(0, ceil(10*multi), ceil(1000*multi)) 
  transformed_load = scipy.stats.norm.cdf(loadspace, loc=0.5*multi, scale=0.1*multi)
  print(loadspace[lim], transformed_load[lim])
  plt.plot(loadspace, transformed_load, label=str(multi))

plt.xlim(0,2)
plt.legend()
plt.xlabel("System Load")
plt.ylabel("Forward Probability")
plt.show()
```
